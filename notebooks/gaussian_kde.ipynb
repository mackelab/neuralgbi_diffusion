```python
import numpy as np
import matplotlib.pyplot as plt

class GaussianKDE:
    def __init__(self, samples: np.ndarray):
        pass
```

```python
samples = []
samples.append(np.random.normal(-1, 0.5, size=(1000, 2)))
samples.append(np.random.normal(1, 0.5, size=(1000, 2)))
samples = np.concatenate(samples, axis=0)

plt.scatter(*samples.T)
```

```python
from sklearn.metrics.pairwise import rbf_kernel

K = rbf_kernel(samples, gamma = 10)
plt.scatter(*samples.T, c=K.sum(axis=0) / K.sum(), s=10)
```
